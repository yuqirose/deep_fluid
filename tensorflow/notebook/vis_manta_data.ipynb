```python
import sys
sys.path.append("../data/sim_1000")

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image

dir_path ="../data/sim_1000/"
```

```python
dat = np.load(dir_path+'velocity_low_0000.npz')
img = dat['arr_0'] # 1x64x64x3
imgplot = plt.imshow(img.squeeze())
```

```python
num_frame = 120
x_dim = 64
y_dim = 64
n_chanel = 3
image_array = np.ndarray((num_frame, x_dim, y_dim, n_chanel))
for i in range(num_frame):
    img_id = str(i).zfill(4)
    image_array[i,:,:,:] = np.load(dir_path+'velocity_low_'+img_id+'.npz')['arr_0']
    
print(image_array.shape)
```

```python
%matplotlib notebook
import matplotlib.animation as animation
plt.rcParams["animation.html"] = "html5"
from IPython.display import HTML

fig = plt.figure()
ims = []
for i in range(num_frame):
    img_id = str(i).zfill(4)
    img = np.load(dir_path+'velocity_low_'+img_id+'.npz')['arr_0'].squeeze()
    im = plt.imshow(img, animated=True)
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval=100, blit=True)
```
